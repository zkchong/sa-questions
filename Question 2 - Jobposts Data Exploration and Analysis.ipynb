```python
import re

def parser_text_to_line(input_str):
    input_str = 'COMPANYNAME: ' + input_str

    input_line = input_str.split('\r\n')
    
    txt_list = []
    cur_txt = ''
    for line in input_line:
        if line.startswith('----------------------------------'):
            break
        elif line[0:3].isupper() and re.match('[A-Z /]+:', line):
            txt_list.append(cur_txt.strip())
            cur_txt = line + ' '
        else:
            cur_txt += line + ' '
        
    txt_list.append(cur_txt.strip())
    return txt_list[1:]
```

```python
def parser_split_lines(line_list):
    new_line_list = []
    for line in line_list:
        assert(line[:3].isupper())
        index = line.index(':')
        new_line = [line[:index].strip(), line[index+1:].strip()]
        new_line_list.append(new_line)
    
    return new_line_list
```

```python

keyword_dict= {
    'COMPANYNAME' : 'company',
    'TITLE' : 'job_title',
    'DURATION' : 'position_duration',
    'PROJECT LOCATION' : 'position_location',
    'DESCRIPTION' : 'job_description',
    'RESPONSIBILITIES' : 'job_responsibility',
    'QUALIFICATION' : 'required_qualification',
    'REMUNERATION' : 'remuneration',
    'SALARY' :  'remuneration',
    'APPLICATION DEADLINE' :'application_deadline',
    'ABOUT' : 'about_company',
}


def parser_to_keyword_dict(line_list):
    result_dict = {}
    #
    for line in line_list:
        for keyword, new_name in zip(keyword_dict.keys(),keyword_dict.values() ):
            if keyword in line[0]:
                result_dict[new_name] = line[1]
                break
    #
    return result_dict
```

e)  Clean text and generate new text from Job Responsibilities column: The new text shall not contain any stop words, and the plural words shall be converted into  singular words. 

```python
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize , RegexpTokenizer
import nltk

stemmer = nltk.PorterStemmer()
stop_words = set(stopwords.words('english')) 

def simplify(input_str):
#     print('input_str:', input_str)
    tokenizer = RegexpTokenizer(r'\w+')    
    word_tokens = tokenizer.tokenize(input_str)
    result_list = [w for w in word_tokens if not w in stop_words]
    #
    result2_list = [ stemmer.stem(r) for r in result_list ]
    return result2_list
```
